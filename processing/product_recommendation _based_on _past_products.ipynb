```python
data1=[]
for doc in data:
    d={}
    d["id"]=doc["id"]
    d["frag"]=doc["frag"]
    s={key:round(value,2) if isinstance(value,float) else value for key,value in doc["colors"].items()}
    d["colors"]=s
    data1.append(d)

data1
```

```python
data = list(map(lambda x:{"id":x["id"],"frag":x["frag"],"colors":{k:v for k,v in x["colors"].items()if k!="Undefined"}},data1))
```

```python
data[0]
```

```python


all_products = data

all_fragrances = ["floral", "rosy", "orchid", "lilies"]

# converting all the products in vectors on basis of their fragnance and coloursall_fragrances = ['rosy', 'floral', 'citrus', 'spicy', 'woody']

# Create feature vectors for clustering
X = []
lengths = []
for product in all_products:
    color_vector = list(product['colors'].values())[0:3]
    frag_vector = [1 if frag in product['frag'] else 0 for frag in all_fragrances]

    # # Ensure both color and fragrance vectors have consistent lengths
    # if len(color_vector) + len(frag_vector) == len(all_fragrances) + len(color_vector):
    feature_vector = color_vector + frag_vector
    feature_vector = [float(val) for val in feature_vector]
    lengths.append(color_vector.__len__())
    X.append(feature_vector)


X = np.array(X)
X
```

```python
X.shape
```

```python
from sklearn.cluster import KMeans


num_clusters = 5
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
kmeans.fit(X)
```

```python
from pprint import pprint
import webbrowser



user_explored_products = [
    {
        "id": "6021b5d0-c12c-4c82-b515-b5ebc04367fe",
        "frag": ["rosy"],
        "colors": {
            "Red": 0.9,
            "Orange": 0.0,
            "Red-Orange": 0.0,
            "Yellow-Orange": 0.0,
        },
    },
    {
        "id": "cbada781-a59d-4f47-b337-8cc6d1567fd5",
        "frag": ["floral"],
        "color": {
            "Yellow-Orange": 0.3578941332199194,
            "Red": 0.28832617328433596,
            "Orange": 0.23322087516665632,
            "Yellow-Green": 0.12055881832908828,
        },
    },
]



frag_list = ["floral", "rosy", "orchid", "lilies"]



colours = list(product["colors"].values())



user_vector = np.mean(
    [

        colours[0:3] + [1 if frag in product["frag"] else 0 for frag in frag_list]
        for product in user_explored_products
    ],

    axis=0,
)


print(user_vector)



# Predict the cluster for the user's feature vector


user_cluster = kmeans.predict([user_vector])[0]



# Get indices of products in the same cluster as the user's products


cluster_indices = np.where(kmeans.labels_ == user_cluster)[0]



# Display or use the recommended products from the same cluster


recommended_products = [
    {
        **all_products[i],
        "link": f"http://localhost:5173/product/{all_products[i]['id']}",
    }
    for i in cluster_indices
][0:10]


pprint(recommended_products)

for product in recommended_products:
    webbrowser.open_new_tab(product["link"])



with open("result.json", "w+") as f:
    import json

    json.dump(recommended_products, f)
```
